```python
def getAddrComp(ip_addr, comp_idx = 0, missing = 'NA'):
    """ Decompose IP Address into 4 components" """
    if not pd.isnull(ip_addr):
        decomps = str(ip_addr).split('.')
        try:
            res = decomps[comp_idx]   
        except:
            res = missing 
    else:
            res = missing 
    return res      
```

```python
## Decompose IP address
df['ip_c1'] = df['ip'].apply(lambda x: getAddrComp(x, comp_idx=0) )
df['ip_c2'] = df['ip'].apply(lambda x: getAddrComp(x, comp_idx=1) )
df['ip_c3'] = df['ip'].apply(lambda x: getAddrComp(x, comp_idx=2) )
df['ip_c4'] = df['ip'].apply(lambda x: getAddrComp(x, comp_idx=3) )
```

```python
## Convert timestamp from string to datetime
df['timestamp'] = pd.to_datetime(df['timestamp'])
```
